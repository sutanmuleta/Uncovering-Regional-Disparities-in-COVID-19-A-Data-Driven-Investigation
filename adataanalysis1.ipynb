```python
import os

# List all files in the extracted directory and their extensions
file_list = []

for root, dirs, files in os.walk('/content/covid19_data'):
    for file in files:
        file_path = os.path.join(root, file)
        file_list.append(file_path)
        print(file_path)

# Check if any CSV files exist
csv_files = [f for f in file_list if f.endswith('.csv')]

if not csv_files:
    print("No CSV files found.")
else:
    print("CSV files found:")
    for csv in csv_files:
        print(csv)
```

```python
import pandas as pd


file_path = '/content/covid19_data/covid-19-open-data-main/src/test/data/hospitalizations.csv'


df = pd.read_csv(file_path)

print(df.head())
```

```python

missing_data = df.isnull().sum()
print("Missing data in each column:")
print(missing_data)
```

```python

df.fillna(0, inplace=True)


print("Missing data after filling:")
print(df.isnull().sum())
```

```python

df['intensive_care_percentage'] = (df['total_intensive_care'] / df['total_hospitalized']) * 100


print(df[['date', 'total_hospitalized', 'total_intensive_care', 'intensive_care_percentage']].head())
```

```python
import matplotlib.pyplot as plt


df.plot(x='date', y='current_hospitalized', title='Current Hospitalizations Over Time')
plt.xlabel('Date')
plt.ylabel('Current Hospitalizations')
plt.show()
```

```python

df.plot(x='date', y='intensive_care_percentage', title='Intensive Care Percentage Over Time')
plt.xlabel('Date')
plt.ylabel('Intensive Care Percentage (%)')
plt.show()
```
